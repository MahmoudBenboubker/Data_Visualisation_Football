```python
def clean_season(path,file_name,competition):
    df = pd.read_csv(path+file_name)
    season = file_name.split('.')[0].split('_')[-1]
    df["id_player"] = df.id
    df.id = df["id"].apply(lambda x : str(x)+'_'+season+'_'+competition)
    positions = ['GK','S','M','D','F']
    for pos in positions:
        df[pos] = df.position.str.contains(pos)*1
    return df
```

```python
clean_season(path,file_name,"laliga")
```

```python
files = [x for x in os.listdir("data/players/la_liga") if "players" in x]
list_dfs = []
for file in files :
    df = clean_season(path,file,"laliga")
    list_dfs.append(df)
```

```python
df_full = pd.concat(list_dfs)
df_full.sort_values("id_player")
```

```python
competition = {"bundesliga" : "bundesliga","epl" : "epl", "la_liga" : "laliga", "ligue_1" : "ligue1", "rfpl" : "rfpl", "serie_a" : "seriea"}

df_final = []
for key, value in competition.items():
    files = [x for x in os.listdir("data/players/"+key) if "players" in x]
    path = "data/players/"+key+'/'
    for file in files :
        df = clean_season(path,file,value)
        df_final.append(df)
df_final = pd.concat(df_final)
```

```python
df_final.to_csv("players.csv", index=False)
```
